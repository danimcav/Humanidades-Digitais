# **Imprimir todos os parágrados do documento que contenham registros de batismos**




````python
import io
from docx import Document

# Obter o nome do arquivo carregado
file_name = next(iter(uploaded))

# Carregar o conteúdo do arquivo Word
file_content = uploaded[file_name]

# Criar um objeto Document do python-docx
doc = Document(io.BytesIO(file_content))

# Variável para rastrear se estamos dentro do parágrafo desejado
found_start = False

# Iterar pelos parágrafos e imprimir a partir do parágrafo que contém "[1]"
for paragraph in doc.paragraphs:
    if "[1]" in paragraph.text:
        found_start = True

    if found_start:
        print(paragraph.text)
````

# **Extrair Nomes das Crianças**

````python
# Variável para rastrear se estamos dentro do parágrafo desejado
found_start = False

# Lista para armazenar palavras iniciadas em maiúscula
nomes = []

# Iterar pelos parágrafos e extrair palavras iniciadas em maiúscula
for paragraph in doc.paragraphs:
    if "[2]" in paragraph.text:
        found_start = True

    if found_start:
        # Extrair palavras iniciadas em maiúscula e adicioná-las à lista
        palavras_maiusculas = [word for word in paragraph.text.split() if word.istitle()]
        nomes.extend(palavras_maiusculas)

# Imprimir a lista de palavras
for nome in nomes:
  print (nome)
````

# **Remover Stopwords**

````python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

# Lista de stopwords em português
stop_words = set(stopwords.words('portuguese'))

# Remover stopwords da lista "nomes"
nomes_sem_stopwords = [nome for nome in nomes if nome.lower() not in stop_words]

# Imprimir a lista de nomes sem stopwords
for nome in nomes_sem_stopwords:
    print(nome)
````

# **Remover Pontuação**

````python
# Remover pontuações da lista "nomes_sem_stopwords"
nomes_sem_stopwords_sem_pontuacao = [nome for nome in nomes_sem_stopwords if nome.isalpha()]

# Imprimir a lista de nomes sem stopwords e sem pontuações
for nome in nomes_sem_stopwords_sem_pontuacao:
    print(nome)
````

````python
import pandas as pd
import re

def find_patronimico_csv(file_path):
    # Lê o arquivo CSV em um dataframe
    df = pd.read_csv(file_path)

    # Expressão regular para identificar o patronímico
    patronimico_regex = r"(\w+)(ez|es|is)$"

    # Encontra o patronímico no sobrenome
    df["patronimico"] = df["Sobrenome"].apply(lambda x: re.search(patronimico_regex, x).group(1) if re.search(patronimico_regex, x) else None)

    # Retorna o dataframe com o patronímico adicionado
    return df
````

````python
import csv

# Abre o arquivo csv
with open('Projeto Integrado I.csv', 'r') as csvfile:

    # Cria um leitor do arquivo
    reader = csv.reader(csvfile, delimiter=',')

    # Inicializa uma lista para armazenar os patronimicos
    patronimicos = []

    # Percorre o arquivo
    for row in reader:

        # Se o nome for composto, o patronímico é o último nome
        if len(row[0].split()) > 1:

            patronimicos.append(row[0].split()[-1])

        # Se o nome for simples, o patronímico é vazio
        else:

            patronimicos.append('')

# Imprime os patronimicos
print(patronimicos)
````
