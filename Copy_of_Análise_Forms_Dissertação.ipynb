### Análise de Sentimento com Modelo Multilingual

```python
from transformers import pipeline

# Load a pre-trained multilingual sentiment analysis model
# You can explore other models on the Hugging Face Hub if this one doesn't meet your needs
sentiment_analyzer = pipeline("sentiment-analysis", model="cardiffnlp/twitter-xlm-roberta-base-sentiment")

sentiment_results_multilingual = {}

qualitative_columns = ['Quais aspectos do seu aprendizado de idiomas você gostaria que fossem mais personalizados?',
                       'Você gostaria de adicionar algum comentário ou sugestão sobre o uso de prompts e IAGen no aprendizado de idiomas?\xa0']

for col in qualitative_columns:
    # Remove missing values and convert to list
    responses = df[col].dropna().tolist()
    if responses:
        # Perform sentiment analysis in batches if needed for performance
        # For this dataset size, processing all at once is likely fine
        results = sentiment_analyzer(responses)
        sentiment_results_multilingual[col] = results
    else:
        sentiment_results_multilingual[col] = []


# Summarize sentiment scores
# The output format of the transformers pipeline is different from VADER
# It usually provides a label (positive, negative, neutral) and a score
for col, results in sentiment_results_multilingual.items():
    if results:
        # Count the occurrences of each sentiment label
        sentiment_counts = {}
        for item in results:
            label = item['label']
            sentiment_counts[label] = sentiment_counts.get(label, 0) + 1

        print(f"Resumo da análise de sentimento para '{col}':")
        for label, count in sentiment_counts.items():
            print(f"  {label}: {count}")
        print("\n")
    else:
        print(f"Nenhuma resposta válida para análise de sentimento em '{col}'")
```

### Filtrar respostas com sentimento negativo ou neutro (Modelo Multilingual)

```python
negative_neutral_responses_multilingual = {}

for col, results in sentiment_results_multilingual.items():
    if results:
        # Get the original responses for this column, aligning with sentiment analysis results
        original_responses = df[col].dropna().tolist()

        # Filter responses based on 'negative' or 'neutral' sentiment label
        filtered_responses = [original_responses[i] for i, item in enumerate(results) if item['label'] in ['negative', 'neutral']]

        if filtered_responses:
            negative_neutral_responses_multilingual[col] = filtered_responses

# Display the negative and neutral sentiment responses
for col, responses in negative_neutral_responses_multilingual.items():
    print(f"Respostas com sentimento negativo ou neutro para '{col}':")
    for response in responses:
        print(f"- {response}")
    print("\n")
```
