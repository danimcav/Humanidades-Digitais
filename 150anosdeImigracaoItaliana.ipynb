Fazer Nuvem de palavras com os sobrenomes mencionados no Forms




```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Substituir os valores NaN por uma string vazia
df['Sobrenomes'] = df['Sobrenomes'].fillna('')

# Definir as cores
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return np.random.choice(["#FFFFFF", "#FF0000"])  # Branco e Vermelho

# Criar a nuvem de palavras
wordcloud = WordCloud(background_color="#008000", width=1000, height=500, color_func=color_func).generate(' '.join(df['Sobrenomes']))

# Plotar a nuvem de palavras
plt.figure(figsize=(15,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

Análise de Sentimentos da coluna Relevancia

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Baixar o léxico de Vader
nltk.download('vader_lexicon')

# Inicializar o analisador de sentimentos
sia = SentimentIntensityAnalyzer()

# Aplicar o analisador de sentimentos à coluna 'Relevancia'
df['sentiment'] = df['Relevancia'].apply(lambda x: sia.polarity_scores(x))

# Visualizar as primeiras linhas do DataFrame
df.head()
```

```python
from textblob import TextBlob

# Aplicar a análise de sentimentos à coluna 'Relevancia'
df['sentiment'] = df['Relevancia'].apply(lambda x: TextBlob(x).sentiment)

# Separar a polaridade e a subjetividade em colunas diferentes
df[['polarity', 'subjectivity']] = df['sentiment'].apply(pd.Series)

# Remover a coluna 'sentiment'
df = df.drop('sentiment', axis=1)

# Visualizar as primeiras linhas do DataFrame
df.head()
```

```python
# Criar uma nova coluna para categorizar os sentimentos
df['sentiment_category'] = df['polarity'].apply(lambda x: 'positivo' if x > 0 else ('neutro' if x == 0 else 'negativo'))

# Criar o histograma
plt.figure(figsize=(8,6))
df['sentiment_category'].value_counts().plot(kind='bar')
plt.title('Distribuição dos Sentimentos')
plt.xlabel('Categoria do Sentimento')
plt.ylabel('Frequência')
plt.show()
```
